## Plot many 'n' column vectors in one graph 

```python
##Box plot
sns.boxplot([df.Summation_IMF1, df.Summation_IMF2, df.Summation_IMF3, df.Summation_IMF4]) # plot 'n' col vecs
```

```python
##Violin plot
sns.violinplot([df.Summation_IMF1, df.Summation_IMF2, df.Summation_IMF3, df.Summation_IMF4]) # plot 'n' col vecs
```

```python
##Heatmap
sns.heatmap([df.Summation_IMF1, df.Summation_IMF2, df.Summation_IMF3, df.Summation_IMF4], annot=True, fmt="d")
```

```python
##Cluster map
sns.clustermap(df)
```
